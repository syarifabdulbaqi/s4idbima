```python
import cv2
```

```python
img_color = cv2.imread('o-img-82363-mobil-berwarna-warni.jpg',1)
img = cv2.imread('o-img-82363-mobil-berwarna-warni.jpg',0)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(img_color) #tampilkan gambar uji coba, ternyata gambarnya tampil dengan format bgr
plt.axis("off")
plt.show()
```

```python
plt.axis("off") #lalu kita ubah dari tampilan bgr ke rgb , agar terlihat seperti gambar semula
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Set thresholding value
threshold_value = 50
```

```python
# Set maximum value
maximum_value = int(str(35)[-3:])  # Ubah "123456789" dengan NIM mahasiswa
```

```python
# Apply thresholding
_, thresholded_image = cv2.threshold(img, threshold_value, maximum_value, cv2.THRESH_BINARY)
```

```python
plt.axis("off") #tampilkan gambar ori format RGB
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
plt.imshow(thresholded_image) #tampilan gambar setelah kita ubah thresholdernya
plt.axis("off")
plt.show()
```
